## String Formatting (f-strings)

F-strings (formatted string literals) provide a way to embed expressions inside string literals, using a minimal syntax. They are denoted by an `f` or `F` prefix before the opening quotation mark or triple quotation mark.

```python
my_string_with_space = "   Hello World   "
print(f"Original string: '{my_string_with_space}'")
print(f"Stripped string: '{my_string_with_space.strip()}'")
```

```python
name = "Alice"
age = 30

# Using f-string for basic variable embedding
print(f"My name is {name} and I am {age} years old.")
```

```python
# You can also embed expressions directly
print(f"In 5 years, I will be {age + 5} years old.")
```

```python
# Formatting numbers
pi = 3.14159
print(f"Pi to two decimal places: {pi:.2f}")
```

The line `print(f"Pi to two decimal places: {pi:.2f}")` uses an f-string to format the `pi` variable.

*   `f"..."`: This indicates an f-string.
*   `{pi:.2f}`: This is the formatting mini-language within the f-string.
    *   `pi`: The variable whose value you want to format.
    *   `:`: Separates the variable from the format specifier.
    *   `.2f`: The format specifier.
        *   `.2`: Specifies that you want to display the number with exactly two digits after the decimal point.
        *   `f`: Specifies that the number should be treated as a fixed-point number (float).

Here are more examples of number formatting with f-strings:

```python
# Aligning text
item = "Apple"
price = 1.25
print(f"{item:<10} {price:>7.2f}") # Left-align item, right-align price
```

```python
value = 123.456789
percentage = 0.75
large_number = 1000000

print(f"Original value: {value}")

# Format to 4 decimal places
print(f"Value to 4 decimal places: {value:.4f}")

# Format as a percentage
print(f"Percentage: {percentage:.2%}")

# Format with thousands separator
print(f"Large number with comma: {large_number:,}")

# Combine with decimal places and thousands separator
print(f"Large number with comma and 2 decimals: {large_number:,.2f}")

# Zero-padding
number = 7
print(f"Zero-padded number (width 3): {number:03d}")
```

The line `print(f"Zero-padded number (width 3): {number:03d}")` uses an f-string to format the `number` variable for zero-padding.

*   `f"..."`: This indicates an f-string.
*   `{number:03d}`: This is the formatting mini-language within the f-string.
    *   `number`: The variable whose value you want to format.
    *   `:`: Separates the variable from the format specifier.
    *   `03d`: The format specifier.
        *   `0`: Specifies that the number should be zero-padded.
        *   `3`: Defines the minimum total width of the output string (including the number itself).
        *   `d`: Specifies that the number should be treated as a decimal integer.
