#### List vs NumPy - Time Taken

````python
from time import process_time
````

**Time taken by a list**

````python
python_list = [i for i in range(10000)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end_time = process_time()

print(end_time - start_time)
````

````python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array += 5

end_time = process_time()

print(end_time - start_time)
````

#### Initial Placeholders in numpy arrays

````python
# create a numpy array of zeros
x = np.zeros((4,5))
print(x)
````

````python
# Create a numpy array of ones
y = np.ones((3,3))
print(y)
````

````python
# create an identity matrix
a = np.eye(5)
print(a)
````

````python
# Create a numpy array with random values
b = np.random.random((3,4))
print(b)
````

#### Randint
**Syntax:**
```python
np.random.randint(Lower_bound, Upper_bound, (rows, columns))

````python
# Create a random integer values array within a specific range
c = np.random.randint(10,100, (3,5))
print(c)
````

````python
dice = np.random.randint(1, 6)
print(dice)
````

````python
# Create a array of evenly spaced values ---> Specifying the number of values required
d = np.linspace(10,30,5)
print(d)
````

````python
# Create a array of evenly spaced values --> Specifying the step
e = np.arange(10,30,5)
print(e)
````

````python
# Convert a list to a numpy array
list_2 = [10,20,30,40,50]

np_array = np.asarray(list_2)
print(np_array)
type(np_array)
````

#### Array Manipulation

````python
array = np.random.randint(0,10, (2,3))
print(array)
print(array.shape)
````

````python
# Transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
````

````python
array = np.random.randint(0,10, (2,3))
print(array)
print(array.shape)
````

````python
trans_2 = array.T
print(trans_2)
print(trans_2.shape)
````

````python
# Reshaping a array
array_2 = np.random.randint(0,10, (2,3))
print(array_2)
print(array_2.shape)
````

````python
array_3 = array_2.reshape(3,2)
print(array_3)
print(array_3.shape)
````
