# Types of Objects in Python:


1.   Immutable Objects
2.   Mutable Objects



Immutable Objects:


1.   int
2.   float
3.   string
4.   bool
5.   tuple

Mutable Objects:
1.  List
2.  Set
3.  Dictionary



````python
# List should be included in the square brackets
my_list = [1,2,3,4,5]
print(my_list)
type(my_list)
````

````python
# Lists can have multiple data types
my_list = [2, 3, 1.8, 'English', True]
print(my_list)
````

Lists are Mutable --> Changeable

````python
# Add elements to a list
my_list = [2, 3, 1.8, 'English', True]
my_list.append(6)
print(my_list)
````

````python
# Print elements of a list using their index
print(my_list[0])
print(my_list[1])
print(my_list[2])
````

````python
# Lists Allow duplicate values
list_1 = [1,2,3,4,5,12,2,3]
print(list_1)
````

````python
print(len(list_1))
````

````python
#  Initiating an empty list
list_2 = []
print(list_2)
````

````python
list_2.append(5)
print(list_2)
````

````python
list_2.append(8)
list_2.append(12)
print(list_2)
````

````python
# delete an item in a list
list_2 = [2, 3, 1.8, 'English', True, 6]
print(list_2)
````

````python
del list_2[2]
print(list_2)
````

````python
# Join two lists
list_3 = [1, 2, 3, 4, 5]
list_4 = [6, 7, 8, 9, 10]

list_5 = list_3 + list_4
print(list_5)
````

# Tuple

````python
tuple_1 = (1,2,3,4,5)
print(tuple_1)
type(tuple_1)
````

````python
# Tuple allows multiple data types
tuple_2 = (1,2,3,4,5, 'Machine Learning', 'Data Science', True)
print(tuple_2)
````

````python
# Converting a list to a tuple

my_list = [1,2,3,4,5]
print(my_list)


my_tuple = tuple(my_list)
print(my_tuple)
````

````python
print(my_tuple[0])
print(my_tuple[1])
print(my_tuple[2])
````

- **Tuples are immutable --> Unchanged**

````python
my_tuple = (1, 2, 3, 4, 5, 'Machine Learning', 'Data Science', True)
print(my_tuple)
my_tuple.append(7)
````

````python
print(len(my_tuple))
````

# Dictionary
A Python dictionary is an unordered collection of data values, used to store data in ``` key:value ``` pairs. Each key must be unique and immutable, while values can be of any data type and can be duplicated.

```
# Key : value
```



````python
my_dictionary = {'name':'Moynul', 'age':25, 'city':'Bangladesh'}
print(my_dictionary)
type(my_dictionary)
````

````python
print(my_dictionary['name'])
print(my_dictionary['age'])
print(my_dictionary['city'])
````

````python
# Dictionary does not allow duplicates values
dictionary_2 = {'name':'Moynul', 'age':25, 'city':'Bangladesh', 'name':'Moynul', 'age':25, 'city':'Bangladesh'}
print(dictionary_2)
````
