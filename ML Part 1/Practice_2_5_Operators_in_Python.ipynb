#### 1. Arithmetic Opearion

```python
num_1 = 20
num_2 = 10

#Addition
sum = num_1 + num_2
print("Summation :", sum)

#Subtraction
diff = num_1 - num_2
print("Difference :", diff)

# Multiplication
prod = num_1 * num_2
print("Product :", prod)

# Division
quot = num_1 / num_1
print("Quotient :", quot)

#Exponent
exp = num_1 ** num_2   # 20^10
print("Exponent :", exp)

#Modulus
mod = num_1 % num_2
print("Reminder :", mod)
```

#### 2. Assignment Operator
- +=
- -=
- *=
- **=
- /=
- %=

```python
a = 5
print(a)
```

```python
a = 5
a +=5  # a = a + 5
print(a)
```

```python
b = 5
b -= 2  # b = b - 2
print(b)
```

```python
c = 5
c *= 2  # c = c * 2
print(c)
```

```python
d = 5
d **= 3  # b = b ^ 3
print(d)
```

```python
e = 5
e /= 2  # e = e / 2
print(e)
```

```python
f = 5
f %= 2  # f = f % 2
print(f)
```

#### 3. Comparison Operator

```python
a = 7
b = 15

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

#### 4. Logical Operators

- and
- or
- not

```python
a = 10

print(a > 20 and a > 5)
print(a > 20 or a > 5)
print(not(a > 8 and a > 5))
```

#### 5. Identity Operators
- is
- is not

```python
x = 5
y = 5

print(x is y)
```

```python
x = 5
y = 10

print(x is y)
```

```python
x = 5
y = 5

print(x is not y)
```

```python
x = 5
y = 10

print(x is not y)
```

#### 6. Membership Operators
- in
- not in

```python
a = 5

c = [1,2,3,4,5]
print(a in c)
```

```python
b = 10

c = [1,2,3,4,5]
print(b in c)
```

```python
a = 5

c = [1,2,3,4,5]
print(a not in c)
```

```python
b = 10

c = [1,2,3,4,5]
print(b not in c)
```
