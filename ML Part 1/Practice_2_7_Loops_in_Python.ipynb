## For Loop
**Syntax:**
```python
for variable in iterable:
    # code block

````python
for i in range(7):    #Note that range(7) is not the values of 0 to 7, but the values 0 to 6.
  laptop_price = int(input("Enter the price  of the laptop : "))
````

````python
num = [10, 20, 30, 40, 50, 60]

print(num[0])
print(num[1])
print(num[2])
print(num[3])
print(num[4])
print(num[5])
````

````python
num = [10, 20, 30, 40, 50, 60]

for i in num:
  print(i)
````

````python
for x in "Banana":
  print(x)
````

````python
fruits = ["Apple", "Banana", "Cherry"]

for x in fruits:
  print(x)
````

````python
fruits = ["Apple", "Banana", "Cherry"]
for x in fruits:
  if x == "Banana":
    break
  print(x)
````

````python
fruits = ["Apple", "Banana", "Cherry"]
for x in fruits:
  if x == "Banana":
    continue
  print(x)
````

#### Else in For Loop

````python
for x in range(6):
  print(x)
else:
  print("Finally finished!")
````

#### Nested Loops

````python
adj = ["Red", "Big", "Tasty"]
fruits = ["Apple", "Banana", "Cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
````

## While Loop

**Syntax:**
```python
while condition:
    # code block

````python
i = 1
while i <= 5:
    print(i)
    i += 1
````

````python
i = 0

while i < 10:
  print(i)
  i += 1
````

````python
# The loop condition (i < 3) is False because i starts at 5,
# so the while loop does not execute even once.

i = 5

while i<3:
  print(i)
  i += 1
````

````python
i = 0
while i < 6:
  print(i)
  i += 1
````

#### The break Statement

````python
i = 0
while i < 6:
  if i == 3:
    break   # stops the loop when i becomes 3
  print(i)
  i += 1
````

````python
i = 1

while i <= 10:
    if i == 6:
        break   # stops the loop when i becomes 6
    print(i)
    i += 1
````

#### The continue Statement

````python
i = 0

while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)

# Note that number 3 is missing in the result
````
